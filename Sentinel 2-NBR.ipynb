```python
import rasterio
```

```python
scr_b08_pre= rasterio.open("2025-08-16-00_00_2025-08-16-23_59_Sentinel-2_L2A_B08_(Raw).tiff")
scr_b012_pre= rasterio.open ("2025-08-16-00_00_2025-08-16-23_59_Sentinel-2_L2A_B12_(Raw).tiff")
scr_b08_post = rasterio.open ("2025-08-23-00_00_2025-08-23-23_59_Sentinel-2_L2A_B08_(Raw).tiff")
scr_b012_post = rasterio.open ("2025-08-23-00_00_2025-08-23-23_59_Sentinel-2_L2A_B12_(Raw).tiff")
```

```python
scr_b08_present = scr_b08_pre.read()[0]
scr_b012_present = scr_b012_pre.read()[0]
scr_b08_post = scr_b08_post.read()[0]
scr_b012_post = scr_b012_post.read()[0]
```

```python
# NBR pre-incendio
dividendo_present = scr_b08_present - scr_b012_present
divisor_present = scr_b08_present + scr_b012_present

# NBR post-incendio
dividendo_post = scr_b08_post - scr_b012_post
divisor_post = scr_b08_post + scr_b012_post

NBR_present = dividendo_present / divisor_present
NBR_post = dividendo_post/ divisor_post
```

```python
NBR_present = dividendo_present / divisor_present
NBR_post = dividendo_post/ divisor_post
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
plt.imshow(NBR_present, cmap='RdYlGn', vmin=0, vmax=1)
plt.colorbar()
plt.title('NBR Present')
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.imshow(NBR_post, cmap='RdYlGn', vmin=0, vmax=1)
plt.colorbar()
plt.title('NBR Post')
plt.show()
```
