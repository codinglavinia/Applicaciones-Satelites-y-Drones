```python
import rasterio
```

```python
scr_false= rasterio.open("2026-06-03-00_00_2026-06-03-23_59_Sentinel-2_L2A_False_color.tiff")
scr_true= rasterio.open ("2026-06-03-00_00_2026-06-03-23_59_Sentinel-2_L2A_True_color.tiff")
```

```python
R,_,_ =scr_true.read()
NIR,_,_= scr_false.read()
```

```python
dividendo = NIR -R
divisor = NIR + R
```

```python
NDVI = dividendo/divisor
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.imshow(NDVI, cmap='RdYlGn', vmin=-1, vmax=1)
plt.colorbar()
plt.title('NDVI')
plt.show()
```
